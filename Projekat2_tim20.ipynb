# Biblioteka i jezicki model za obradu teksta

```python
!pip -q install transformers torch
from transformers import pipeline
import random

llm = pipeline(
    task="text-generation",
    model="distilgpt2",
    max_new_tokens=150
)
```

# Poizvanje LLM modela

```python
def call_llm(prompt):
    try:
        return llm(prompt)[0]["generated_text"]
    except Exception as e:
        return f"error {e}"
```

# Simulira se pretraga naucne literature

```python
def arxiv_search_tool(query, max=3):
    if query is None or query == "":
        raise ValueError("Query ne sme da bude prazan")

    baza = [
        {
            "title": "Learning with Noisy Labels",
            "authors": ["Natarajan et al."],
            "year": 2013,
            "abstract": "Classification under label noise."
        },
        {
            "title": "Outlier Detection Techniques",
            "authors": ["Chandola et al."],
            "year": 2009,
            "abstract": "Comprehensive review of outlier detection."
        },
        {
            "title": "Data Cleaning for Machine Learning",
            "authors": ["Rahm et al."],
            "year": 2000,
            "abstract": "Survey of data cleaning techniques."
        }
    ]

    return random.sample(baza, min(max, len(baza)))
```

# Definisemo klasu agenta

```python
class Agent:
    def __init__(self, name):
        self.name = name
        self.memory = {}

    def log(self, message):
        print(f"[{self.name}] {message}")
```

# Agent analizira upit i pravi plan researcha

```python
class ResearchPlannerAgent(Agent):
    def run(self, user_query):
        self.log("Analiza zahteva")

        plan = {
            "main_topic": user_query,
            "subtopics": [
                "data cleaning",
                "outlier detection",
                "label noise"
            ],
            "keywords": [
                user_query,
                "noisy data",
                "machine learning"
            ]
        }

        self.memory["plan"] = plan
        return plan
```

# Agent koristi alat za pronalazenje relevantnih naucnih radova

```python
class LiteratureSearchAgent(Agent):
    def run(self, plan):
        self.log("Trazenje literature")

        seen_titles = set()
        papers = []

        for keyword in plan["keywords"]:
            try:
                results = arxiv_search_tool(keyword)
                for paper in results:
                    if paper["title"] not in seen_titles:
                        seen_titles.add(paper["title"])
                        papers.append(paper)
            except Exception as e:
                self.log(f"error {e}")

        self.memory["papers"] = papers
        return papers
```

# Agent primenjuje unapred definisanu rutinu za analizu svakog naucnog rada

```python
class PaperAnalysisAgent(Agent):
    def run(self, papers):
        self.log("Analiziram prikupljene radove")

        analize = []

        for paper in papers:
            self.log(f"Analyzing: {paper['title']}")

            analiza = {
                "METAPODACI": {
                    "naslov": paper["title"],
                    "autori": paper["authors"],
                    "godina": paper["year"],
                    "oblast": "Machine Learning"
                },
                "CILJ RADA": "Poboljsanje rada sa losim podacima",
                "METODOLOGIJA": "Pregled i analiza metoda koje vec postoje",
                "REZULTATI": paper["abstract"],
                "OGRANICENJA": "Zavisnost od kvaliteta ulaza",
                "DOPRINOS": "Pregled kljucnih tehnika",
                "OTVORENA PITANJA": "Automatizacija procesa"
            }

            analize.append(analiza)

        self.memory["analyses"] = analize
        return analize
```

# Agent spaja analizirane radove u naucni tekst

```python
class ScientificWriterAgent(Agent):
    def run(self, analyses):
        self.log("Ispisujemo naucni tekst")

        text = "Prikupljeno:\n\n"

        for a in analyses:
            text += (
                f"{a['METAPODACI']['naslov']} "
                f"({a['METAPODACI']['godina']}) "
                f"bavi se temom: {a['REZULTATI']}\n\n"
            )

        self.memory["text"] = text
        return text
```

Agent prikazuje generisani tekst

```python
class InteractionAgent(Agent):
    def run(self, text):
        self.log("Ispis rezultata korisniku")

        print("\nGenerisani tekst\n")
        print(text)
```

# Upravljanje redosledom rada agenata

```python
def run_system(user_query):
    planner = ResearchPlannerAgent("Agent 1 - Planer")
    searcher = LiteratureSearchAgent("Agent 2 - Search")
    analyzer = PaperAnalysisAgent("Agent 3 - Analiza")
    writer = ScientificWriterAgent("Agent 4 - Writer")
    interactor = InteractionAgent("Agent 5 - Interakcija")

    plan = planner.run(user_query)
    papers = searcher.run(plan)
    analyses = analyzer.run(papers)
    text = writer.run(analyses)
    interactor.run(text)
```

# Pokretanje naseg sistema

```python
run_system("Optimizacija skupa podataka loseg kvaliteta")
```
